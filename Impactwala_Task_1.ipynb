```python
from urllib import request
```

```python
#getting the url
data ="https://rankabrand.org/electronics/Apple"
```

```python
#reading url
page = urllib.request.urlopen(data)
```

```python
soup =  BeautifulSoup(page)
```

```python
#applying prettify to get HTML text
print (soup.prettify())
```

```python
#viewing title
soup.title
```

```python
#To view table body
soup.tbody
```

```python
right_table=soup.findAll('table', class_="score-report")
right_table
```

```python
right_table1 = right_table[0]
right_table2 = right_table[1]
right_table3 = right_table[2]
```

```python
A=[]
B=[]
C=[]
D=[]
for row in right_table1.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') 
    A.append(cells[0].find(text=True))
    B.append(cells[1].find(text=False))
    C.append(cells[2].find(text=True))
    D.append(cells[3].find(text=False))

 
import pandas as pd
df1=pd.DataFrame(A,columns=['Question'])
df1['Status']=B
df1['Remark']=C
df1['Source']=D
df1       
```

```python
A=[]
B=[]
C=[]
D=[]
for row in right_table2.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') 
    A.append(cells[0].find(text=True))
    B.append(cells[1].find(text=False))
    C.append(cells[2].find(text=True))
    D.append(cells[3].find(text=False))
  

import pandas as pd
df2=pd.DataFrame(A,columns=['Question'])
df2['Status']=B
df2['Remark']=C
df2['Source']=D
df2
```

```python
A=[]
B=[]
C=[]
D=[]
for row in right_table3.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th')
    A.append(cells[0].find(text=True))
    B.append(cells[1].find(text=False))
    C.append(cells[2].find(text=True))
    D.append(cells[3].find(text=False))
    
    
import pandas as pd
df3=pd.DataFrame(A,columns=['Question'])
df3['Status']=B
df3['Remark']=C
df3['Source']=D
df3   
```

```python
#To concatenate the three extrated values
df4 = pd.concat([df1,df2,df3]).to_csv("Apple_data.csv", header = True, encoding = 'utf-8')
```
